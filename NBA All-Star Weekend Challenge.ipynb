# What draft class is represeted the most during this time?

```python
df["Draft_Class"] = df["NBA Draft Status"].str.split(" ", expand = False) 
```

```python
df.head()
```

```python
df[["drft_yr", "rnd", "rnd_num","pck", "pck_num"]] = pd.DataFrame(df.Draft_Class.tolist(), index= df.index)
```

```python
df.head()
```

```python
df.drop(['Draft_Class','rnd', 'pck'], axis=1)
```

```python
draft_class_rep = df["drft_yr"].mode()
```

```python
draft_class_rep
```

#### 1996 is represented the most 

# Are foriegn players more prevalent in the All-Star Game during this time?

```python
df["foreign_or_not"] = np.where(df["Nationality"]=="United States", "US", "No")
```

```python
df
```

```python
foreign = df['foreign_or_not'].value_counts()
foreign
```

#### Most players are from the US (365) vs foreign (74)

# How often are the Western All-Stars voted in by fans?

```python
west_fans = df['Selection Type'].value_counts()
west_fans 
```

# How many times does Steph Curry make the All-Star game during this time period?

```python
steph_freq = df[(df.Player=="Stephen Curry")].count()
steph_freq 
```
